```python
import zipfile
import os
import dicom2nifti

def convert_zip_to_nifti(zip_filepath):
    # Extract the zip file
    with zipfile.ZipFile(zip_filepath, 'r') as zip_ref:
        extract_dir = os.path.splitext(zip_filepath)[0]
        zip_ref.extractall(extract_dir)
    
    # Convert DICOM to NIfTI
    dicom2nifti.convert_directory(extract_dir, extract_dir)
    
    # Rename the output file to match the zip file name
    nifti_filename = os.path.basename(zip_filepath).replace('.zip', '.nii.gz')
    nifti_filepath = os.path.join(extract_dir, "..", nifti_filename)
    
    # Move the NIfTI file to the same directory as the zip file
    os.rename(os.path.join(extract_dir, 'none.nii.gz'), nifti_filepath)


    
    return nifti_filepath

# Example usage
# nifti_file = convert_zip_to_nifti('path_to_your_zip_file.zip')
```

```python
import SimpleITK as sitk
import pandas as pd
import numpy as np
import os
from skimage import measure

def count_disconnected_regions_simple(vessel_image_path):
    # Read the vessel image
    vessel_image = sitk.ReadImage(vessel_image_path)
        
    # Convert the image to a numpy array
    vessel_array = sitk.GetArrayFromImage(vessel_image)
        
    # Label the connected regions
    _, num_regions = measure.label(vessel_array, return_num=True, connectivity=3)
        
    return num_regions

def analyze_images(subfolder_path, suffix):
    """
    Analyze the images in the given subfolder and return the statistics as a dictionary.
    
    Parameters:
    - subfolder_path: str, the path to the subfolder containing the images
    - suffix: str, the suffix of the image files (e.g., '_cropped.nii.gz')
    
    Returns:
    - result: dict, the statistics of the images
    """
    patient_nifti_path = os.path.join(subfolder_path, f'patient{suffix}')
    liver_nifti_path = os.path.join(subfolder_path, f'liver{suffix}')
    potalvein_nifti_path = os.path.join(subfolder_path, f'portalvein{suffix}')
    venacava_nifti_path = os.path.join(subfolder_path, f'venacava{suffix}')
    
    if os.path.exists(patient_nifti_path) and os.path.exists(liver_nifti_path) and os.path.exists(potalvein_nifti_path) and os.path.exists(venacava_nifti_path):
        # Read the NIfTI files
        patient_image = sitk.ReadImage(patient_nifti_path)
        liver_image = sitk.ReadImage(liver_nifti_path)
        potalvein_image = sitk.ReadImage(potalvein_nifti_path)
        venacava_image = sitk.ReadImage(venacava_nifti_path)
        
        # Get spacing and size
        patient_spacing = patient_image.GetSpacing()
        liver_spacing = liver_image.GetSpacing()
        potalvein_spacing = potalvein_image.GetSpacing()
        venacava_spacing = venacava_image.GetSpacing()

        # Get orientation
        patient_orientation = patient_image.GetDirection()
        liver_orientation = liver_image.GetDirection()
        potalvein_orientation = potalvein_image.GetDirection()
        venacava_orientation = venacava_image.GetDirection()

        # Calculate the range of pixel values in the patient image
        patient_array = sitk.GetArrayFromImage(patient_image)
        pixel_min = np.min(patient_array)
        pixel_max = np.max(patient_array)
        pixel_range = (pixel_min, pixel_max)

        # Calculate unique values in liver and vessels images
        liver_array = sitk.GetArrayFromImage(liver_image)
        liver_unique_values = np.unique(liver_array)
        
        portalvein_array = sitk.GetArrayFromImage(potalvein_image)
        portalvein_unique_values = np.unique(portalvein_array)

        venacava_array = sitk.GetArrayFromImage(venacava_image)
        venacava_unique_values = np.unique(venacava_array)
        
        patient_size = patient_image.GetSize()
        liver_size = liver_image.GetSize()
        potalvein_size = potalvein_image.GetSize()
        venacava_size = venacava_image.GetSize()

        # Check if spacing and size are the same
        if patient_spacing != liver_spacing or patient_spacing != potalvein_spacing or patient_spacing != venacava_spacing:
            raise ValueError(f"Spacing mismatch in subfolder {subfolder_path}")
        if patient_size != liver_size or patient_size != potalvein_size or patient_size != venacava_size:
            raise ValueError(f"Size mismatch in subfolder {subfolder_path}")
        # Check if orientations are the same
        if patient_orientation != liver_orientation or patient_orientation != potalvein_orientation or patient_orientation != venacava_orientation:
            raise ValueError(f"Orientation mismatch in subfolder {subfolder_path}")

        # Calculate length
        length = tuple(p * s for p, s in zip(patient_spacing, patient_size))

        # Count disconnected regions in portalvein and venacava images
        # Label the connected regions directly from the numpy arrays
        _, portalvein_disconnected_regions = measure.label(portalvein_array, return_num=True,  connectivity=3)
        _, venacava_disconnected_regions = measure.label(venacava_array, return_num=True,  connectivity=3)

        # Record the results
        result = {
            'Subfolder': os.path.basename(subfolder_path),
            'Spacing': list(patient_spacing),
            'Size': list(patient_size),
            'Length': list(length),
            'orientation': list(patient_orientation),
            'Pixel Range': list(pixel_range),
            'Liver Unique Values': list(liver_unique_values),
            'Portalvein Unique Values': list(portalvein_unique_values),
            'Venacava Unique Values': list(venacava_unique_values),
            'Portalvein Disconnected Regions': portalvein_disconnected_regions,
            'Venacava Disconnected Regions': venacava_disconnected_regions,
        }
        
        return result
    else:
        raise FileNotFoundError("One or more required files are missing.")
```

```python

def analyze_and_save_images(folder_root, suffix, excel_filename):
    """
    Traverse all subfolders in folder_root, analyze the images, and save the results to an Excel file.
    
    Parameters:
    - folder_root: str, the root folder containing the subfolders with images
    - suffix: str, the suffix of the image files (e.g., '.nii.gz')
    - excel_filename: str, the name of the output Excel file
    """
    results = []
    for subfolder in os.listdir(folder_root):
        subfolder_path = os.path.join(folder_root, subfolder)
        
        if os.path.isdir(subfolder_path):
            try:
                result = analyze_images(subfolder_path, suffix)
                results.append(result)
            except Exception as e:
                print(f"Error processing {subfolder_path}: {e}")

    # Create a DataFrame from the results
    df_results = pd.DataFrame(results)

    # Write the results to an Excel file
    output_excel_path = os.path.join(folder_root, excel_filename)
    df_results.to_excel(output_excel_path, index=False)
    print(f"Results saved to {output_excel_path}")

# Example usage
# analyze_and_save_images(folder_root, '.nii.gz', 'original_image_analysis.xlsx')
```

```python
import SimpleITK as sitk
import numpy as np

def crop_nifti_images(patient_path, liver_path, portalvein_path, venacava_path):
    # Read the liver NIfTI image to get the organ region coordinates
    liver_image = sitk.ReadImage(liver_path)
    liver_array = sitk.GetArrayFromImage(liver_image)
    non_zero_indices = np.argwhere(liver_array != 0)
    x_min, y_min, z_min = map(int, non_zero_indices.min(axis=0))
    x_max, y_max, z_max = map(int, non_zero_indices.max(axis=0))

    mask = liver_array[x_min:x_max+1, y_min:y_max+1, z_min:z_max+1]

    # Function to crop and save the image
    def crop_and_save(image_path):
        image = sitk.ReadImage(image_path)
        image_array = sitk.GetArrayFromImage(image)
        cropped_image_array = image_array[x_min:x_max+1, y_min:y_max+1, z_min:z_max+1]
        cropped_image_array[mask==0] = np.min(image_array)
        cropped_image = sitk.GetImageFromArray(cropped_image_array)
        cropped_image.SetDirection(image.GetDirection())
        cropped_image.SetSpacing(image.GetSpacing())
        cropped_image.SetOrigin(image.TransformIndexToPhysicalPoint([x_min, y_min, z_min]))
        for key in image.GetMetaDataKeys():
            cropped_image.SetMetaData(key, image.GetMetaData(key))
        output_path = image_path.replace('.nii.gz', '_cropped.nii.gz')
        sitk.WriteImage(cropped_image, output_path)
        print(f'Cropped image saved to {output_path}')

    # Crop and save the patient, liver, portalvein, and venacava images
    crop_and_save(patient_path)
    crop_and_save(liver_path)
    crop_and_save(portalvein_path)
    crop_and_save(venacava_path)

# # Example usage
# patient_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.2/PATIENT_DICOM.nii.gz'
# liver_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.2/liver.nii.gz'
# vessel_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.2/liver_vessels.nii.gz'
# crop_nifti_images(patient_path, liver_path, vessel_path)
```

```python
import SimpleITK as sitk
import numpy as np
import os

def process_images(patient_image_path, liver_image_path, potalvein_image_path, venacava_image_path):
    # Read the images
    patient_image = sitk.ReadImage(patient_image_path)
    liver_image = sitk.ReadImage(liver_image_path)
    potalvein_image = sitk.ReadImage(potalvein_image_path)
    venacava_image = sitk.ReadImage(venacava_image_path)
    
    # Convert images to numpy arrays
    patient_array = sitk.GetArrayFromImage(patient_image)
    liver_array = sitk.GetArrayFromImage(liver_image)
    potalvein_array = sitk.GetArrayFromImage(potalvein_image)
    venacava_array = sitk.GetArrayFromImage(venacava_image)
    
    # Clip the patient image pixel values to the range [-110, 250]
    patient_array = np.clip(patient_array, -110, 250)
    
    # Normalize the clipped patient image to the range [0, 1]
    patient_array = (patient_array - (-110)) / (250 - (-110))
    
    # Set all non-zero pixels in the liver, potalvein, and venacava images to 1
    liver_array[liver_array != 0] = 1
    potalvein_array[potalvein_array != 0] = 1
    venacava_array[venacava_array != 0] = 1
    
    # Convert the numpy arrays back to SimpleITK images
    processed_patient_image = sitk.GetImageFromArray(patient_array)
    processed_liver_image = sitk.GetImageFromArray(liver_array)
    processed_potalvein_image = sitk.GetImageFromArray(potalvein_array)
    processed_venacava_image = sitk.GetImageFromArray(venacava_array)
    
    # Copy the metadata from the original images
    processed_patient_image.CopyInformation(patient_image)
    processed_liver_image.CopyInformation(liver_image)
    processed_potalvein_image.CopyInformation(potalvein_image)
    processed_venacava_image.CopyInformation(venacava_image)
    
    # Generate output paths with '_clip' suffix
    output_patient_path = patient_image_path.replace('.nii.gz', '_clip.nii.gz')
    output_liver_path = liver_image_path.replace('.nii.gz', '_clip.nii.gz')
    output_potalvein_path = potalvein_image_path.replace('.nii.gz', '_clip.nii.gz')
    output_venacava_path = venacava_image_path.replace('.nii.gz', '_clip.nii.gz')
    
    # Save the processed images
    sitk.WriteImage(processed_patient_image, output_patient_path)
    sitk.WriteImage(processed_liver_image, output_liver_path)
    sitk.WriteImage(processed_potalvein_image, output_potalvein_path)
    sitk.WriteImage(processed_venacava_image, output_venacava_path)
    print(f"Processed images saved as {output_patient_path}, {output_liver_path}, {output_potalvein_path}, and {output_venacava_path}")

# Traverse all subfolders in folder_root and process the images
for subfolder in os.listdir(folder_root):
    subfolder_path = os.path.join(folder_root, subfolder)
    
    if os.path.isdir(subfolder_path):
        patient_image_path = os.path.join(subfolder_path, 'patient_cropped.nii.gz')
        liver_image_path = os.path.join(subfolder_path, 'liver_cropped.nii.gz')
        potalvein_image_path = os.path.join(subfolder_path, 'portalvein_cropped.nii.gz')
        venacava_image_path = os.path.join(subfolder_path, 'venacava_cropped.nii.gz')
        
        if os.path.exists(patient_image_path) and os.path.exists(liver_image_path) and os.path.exists(potalvein_image_path) and os.path.exists(venacava_image_path):
            process_images(patient_image_path, liver_image_path, potalvein_image_path, venacava_image_path)
```

```python
import SimpleITK as sitk
import numpy as np
import os

def crop_image(image_path, output_path, crop_dims):
    """
    Crop the given image along specified dimensions and save the result.
    
    Parameters:
    - image_path: str, the path to the input image
    - output_path: str, the path to save the cropped image
    - crop_dims: dict, the dimensions to crop and the amount to crop in each dimension
                 e.g., {'x': (10, 20), 'y': (5, 15), 'z': (0, 10)}
                 This means crop 10 pixels from the start and 20 pixels from the end in the x dimension,
                 5 pixels from the start and 15 pixels from the end in the y dimension,
                 and 0 pixels from the start and 10 pixels from the end in the z dimension.
    """
    # Read the image
    image = sitk.ReadImage(image_path)
    image_array = sitk.GetArrayFromImage(image)
    
    # Get the original size
    original_size = image_array.shape
    
    # Calculate the crop slices
    crop_slices = []
    for dim, (start_crop, end_crop) in crop_dims.items():
        if dim == 'x':
            crop_slices.append(slice(start_crop, original_size[0] - end_crop))
        elif dim == 'y':
            crop_slices.append(slice(start_crop, original_size[1] - end_crop))
        elif dim == 'z':
            crop_slices.append(slice(start_crop, original_size[2] - end_crop))
    
    # Crop the image array
    cropped_array = image_array[crop_slices[0], crop_slices[1], crop_slices[2]]
    
    # Convert back to SimpleITK image
    cropped_image = sitk.GetImageFromArray(cropped_array)
    cropped_image.SetDirection(image.GetDirection())
    cropped_image.SetSpacing(image.GetSpacing())
    
    # Calculate new origin
    original_origin = np.array(image.GetOrigin())
    original_spacing = np.array(image.GetSpacing())
    new_origin = original_origin + np.array([crop[0] for crop in crop_dims.values()]) * original_spacing
    cropped_image.SetOrigin(tuple(new_origin))
    
    # Save the cropped image
    sitk.WriteImage(cropped_image, output_path)
    print(f'Cropped image saved to {output_path}')

# #Example usage
# input_image_path = os.path.join(folder_root, "3Dircadb1.7/patient_cropped_clip.nii.gz")
# output_image_path = os.path.join(folder_root, "3Dircadb1.7/patient_cropped_clip.nii.gz")

# crop_image(input_image_path, output_image_path, {'x': (0, 0), 'y': (0, 0), 'z': (0, 100)})

# #Example usage
# input_image_path = os.path.join(folder_root, "3Dircadb1.12/patient_cropped_clip.nii.gz")
# output_image_path = os.path.join(folder_root, "3Dircadb1.12/patient_cropped_clip.nii.gz")

# crop_image(input_image_path, output_image_path, {'x': (72, 0), 'y': (0, 0), 'z': (0, 0)})
```

```python
import SimpleITK as sitk
import numpy as np
import os

def resample_image(image, new_spacing, is_label=False):
    """
    Resample the given image to the new spacing and size.
    
    Parameters:
    - image: SimpleITK.Image, the image to be resampled
    - new_spacing: tuple, the desired voxel spacing (x, y, z)
    - new_size: tuple, the desired image size (x, y, z)
    - is_label: bool, whether the image is a label image (use nearest neighbor interpolation)
    
    Returns:
    - resampled_image: SimpleITK.Image, the resampled image
    """
    original_spacing = image.GetSpacing()
    original_size = image.GetSize()
    
    # Calculate the new size based on the new spacing
    new_size = [int(np.round(osz * ospc / nspc)) for osz, ospc, nspc in zip(original_size, original_spacing, new_spacing)]
    
    # Define the resampling filter
    resample = sitk.ResampleImageFilter()
    resample.SetOutputSpacing(new_spacing)
    resample.SetSize(new_size)
    resample.SetOutputDirection(image.GetDirection())
    resample.SetOutputOrigin(image.GetOrigin())
    
    # Set the interpolation method
    if is_label:
        resample.SetInterpolator(sitk.sitkNearestNeighbor)
    else:
        resample.SetInterpolator(sitk.sitkLinear)
    
    # Perform the resampling
    resampled_image = resample.Execute(image)
    
    return resampled_image

def crop_and_pad_image(image, new_size):
    """
    Crop and pad the given image to the new size, keeping the original image centered.
    
    Parameters:
    - image: SimpleITK.Image, the image to be cropped and padded
    - new_size: tuple, the desired image size (x, y, z)
    
    Returns:
    - cropped_padded_image: SimpleITK.Image, the cropped and padded image
    """
    image_array = sitk.GetArrayFromImage(image)
    original_size = image_array.shape
    min_pixel_value = np.min(image_array)
    
    # Calculate cropping sizes
    crop_sizes = [(orig - new) // 2 for orig, new in zip(original_size, new_size)]
    crop_slices = [slice(max(0, crop), min(orig, orig - crop)) for crop, orig in zip(crop_sizes, original_size)]
    
    # Crop the image array
    cropped_array = image_array[crop_slices[0], crop_slices[1], crop_slices[2]]
    
    # Calculate padding sizes
    pad_sizes = [(new - cropped) // 2 for new, cropped in zip(new_size, cropped_array.shape)]
    pad_widths = [(pad, new - cropped - pad) for pad, new, cropped in zip(pad_sizes, new_size, cropped_array.shape)]
    
    # Pad the cropped image array
    padded_array = np.pad(cropped_array, pad_widths, mode='constant', constant_values=min_pixel_value)
    
    # Convert back to SimpleITK image
    cropped_padded_image = sitk.GetImageFromArray(padded_array)
    cropped_padded_image.SetDirection(image.GetDirection())
    cropped_padded_image.SetSpacing(image.GetSpacing())
    
    # Calculate new origin
    original_origin = np.array(image.GetOrigin())
    original_spacing = np.array(image.GetSpacing())
    new_origin = original_origin - np.array([crop for crop in crop_sizes]) * original_spacing
    cropped_padded_image.SetOrigin(tuple(new_origin))
    
    return cropped_padded_image

def resample_images_in_folder(folder_path, new_spacing, new_size):
    """
    Resample the patient, liver, portal vein, and vena cava images in the given folder to the new spacing and size.
    
    Parameters:
    - folder_path: str, the path to the folder containing the images
    - new_spacing: tuple, the desired voxel spacing (x, y, z)
    - new_size: tuple, the desired image size (x, y, z)
    """
    # Define the file paths
    patient_image_path = os.path.join(folder_path, 'patient_cropped_clip.nii.gz')
    liver_image_path = os.path.join(folder_path, 'liver_cropped_clip.nii.gz')
    portalvein_image_path = os.path.join(folder_path, 'portalvein_cropped_clip.nii.gz')
    venacava_image_path = os.path.join(folder_path, 'venacava_cropped_clip.nii.gz')
    
    # Read the images
    patient_image = sitk.ReadImage(patient_image_path)
    liver_image = sitk.ReadImage(liver_image_path)
    portalvein_image = sitk.Cast(sitk.ReadImage(portalvein_image_path), sitk.sitkFloat32)
    venacava_image = sitk.Cast(sitk.ReadImage(venacava_image_path), sitk.sitkFloat32)
    
    # Resample the images
    resampled_patient_image = resample_image(patient_image, new_spacing, is_label=False)
    resampled_liver_image = resample_image(liver_image, new_spacing, is_label=True)
    resampled_portalvein_image = resample_image(portalvein_image, new_spacing, is_label=False)
    resampled_venacava_image = resample_image(venacava_image, new_spacing, is_label=False)

    # Crop and pad the resampled images to the new size
    resampled_patient_image = crop_and_pad_image(resampled_patient_image, new_size)
    resampled_liver_image = crop_and_pad_image(resampled_liver_image, new_size)
    resampled_portalvein_image = crop_and_pad_image(resampled_portalvein_image, new_size)
    resampled_venacava_image = crop_and_pad_image(resampled_venacava_image, new_size)
    
    # Clip the portal vein and vena cava images
    portalvein_array = sitk.GetArrayFromImage(resampled_portalvein_image)
    venacava_array = sitk.GetArrayFromImage(resampled_venacava_image)
    
    portalvein_array[portalvein_array > 0.2] = 1
    venacava_array[venacava_array > 0.2] = 1
    
    resampled_portalvein_image_new = sitk.GetImageFromArray(portalvein_array.astype(np.uint8))
    resampled_venacava_image_new = sitk.GetImageFromArray(venacava_array.astype(np.uint8))
    
    # Set the metadata for the clipped images
    resampled_portalvein_image_new.SetDirection(resampled_portalvein_image.GetDirection())
    resampled_portalvein_image_new.SetSpacing(resampled_portalvein_image.GetSpacing())
    resampled_portalvein_image_new.SetOrigin(resampled_portalvein_image.GetOrigin())
    
    resampled_venacava_image_new.SetDirection(resampled_venacava_image.GetDirection())
    resampled_venacava_image_new.SetSpacing(resampled_venacava_image.GetSpacing())
    resampled_venacava_image_new.SetOrigin(resampled_venacava_image.GetOrigin())
    
    # Save the resampled images
    sitk.WriteImage(resampled_patient_image, patient_image_path.replace('.nii.gz', '_resampled.nii.gz'))
    sitk.WriteImage(resampled_liver_image, liver_image_path.replace('.nii.gz', '_resampled.nii.gz'))
    sitk.WriteImage(resampled_portalvein_image_new, portalvein_image_path.replace('.nii.gz', '_resampled.nii.gz'))
    sitk.WriteImage(resampled_venacava_image_new, venacava_image_path.replace('.nii.gz', '_resampled.nii.gz'))
    print(f"Resampled images saved in {folder_path}")

# # Example usage
# patient_image_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.3/PATIENT_DICOM_cropped_clip.nii.gz'
# liver_image_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.3/liver_cropped_clip.nii.gz'
# vessel_image_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.3/liver_vessels_cropped.nii.gz'
# new_spacing = (0.782, 0.782, 1.6)
# new_size = (160, 352, 352)

# resample_images(patient_image_path, liver_image_path, vessel_image_path, new_spacing, new_size)
```

```python
from skimage import measure
import SimpleITK as sitk

def count_and_delete_disconnected_regions(vessel_image_path):
    # Read the vessel image
    vessel_image = sitk.ReadImage(vessel_image_path)
    
    # Convert the image to a numpy array
    vessel_array = sitk.GetArrayFromImage(vessel_image)
    
    # Label the connected regions
    labeled_array, num_regions = measure.label(vessel_array, return_num=True, connectivity=3)
    
    # Calculate the number of pixels in each region
    region_props = measure.regionprops(labeled_array)
    region_pixel_counts = [region.area for region in region_props]
    
    # Remove regions with fewer than 100 pixels
    for region in region_props:
        if region.area < 100:
            labeled_array[labeled_array == region.label] = 0
    
    # Convert the labeled array back to the original vessel array
    vessel_array[labeled_array == 0] = 0
    
    # Convert the modified array back to a SimpleITK image
    modified_vessel_image = sitk.GetImageFromArray(vessel_array)
    modified_vessel_image.CopyInformation(vessel_image)
    
    # Save the modified image
    sitk.WriteImage(modified_vessel_image, vessel_image_path.replace('.nii.gz', '_modified.nii.gz'))
    
    # Recalculate the number of regions after removal
    _, num_regions_after_removal = measure.label(vessel_array, return_num=True, connectivity=3)
    
    return num_regions, num_regions_after_removal, region_pixel_counts

# # Example usage
# vessel_image_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.1/vessels_padded.nii.gz'
# num_regions, num_regions_after_removal, region_pixel_counts = count_disconnected_regions(vessel_image_path)
# print(f'Number of disconnected regions: {num_regions}')
# print(f'Region pixel counts: {region_pixel_counts}')
# print(f'Number of disconnected regions after removal: {num_regions_after_removal}')
```

```python
import SimpleITK as sitk
import numpy as np
from skimage.filters import sato

def process_image_with_sato(input_image_path, output_image_path):
    """
    Process the input image using skimage.filters.sato and save the result as a NIfTI file.
    
    Parameters:
    - input_image_path: str, path to the input NIfTI image
    - output_image_path: str, path to save the processed NIfTI image
    """
    # Read the input image
    input_image = sitk.ReadImage(input_image_path)
    
    # Convert the image to a numpy array
    input_array = sitk.GetArrayFromImage(input_image)
    
    # Apply the Sato filter
    processed_array = sato(input_array, sigmas=range(1, 4, 1), black_ridges=False)
    
    # Convert the processed array back to a SimpleITK image
    processed_image = sitk.GetImageFromArray(processed_array)
    
    # Copy the metadata from the original image
    processed_image.CopyInformation(input_image)
    
    # Save the processed image
    sitk.WriteImage(processed_image, output_image_path)
    print(f"Processed image saved to {output_image_path}")

# # Example usage
# input_image_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.1/patient_adjusted.nii.gz'
# output_image_path = '/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.1/patient_sato.nii.gz'
# process_image_with_sato(input_image_path, output_image_path)
```

```python
import edt
import torch

def label2distance(seg):
    #input: c*z*y*x
    if type(seg) == torch.Tensor:
        seg= seg.numpy()

    posmask = seg> 0
    negmask = ~posmask

    dis_background = edt.edt(negmask)
    dis_foreground = edt.edt(posmask) - 1
        
    dist = dis_background * negmask - dis_foreground * posmask

    return dist
```

```python
import SimpleITK as sitk
import numpy as np

def extract_and_save_liver_contour(liver_image_path):
    """
    Extract the liver contour from the binary segmentation label using Morphological Gradient filter and save it to a new image.
    
    Parameters:
    - liver_image_path: str, path to the liver binary segmentation label image
    """
    # Read the liver image
    liver_image = sitk.ReadImage(liver_image_path)
    
    # Apply Morphological Gradient filter to extract edges
    morph_gradient_filter = sitk.MorphologicalGradientImageFilter()
    contour_image = morph_gradient_filter.Execute(liver_image)
    
    # Binarize the contour image
    contour_array = sitk.GetArrayFromImage(contour_image)
    contour_array = (contour_array > 0).astype(np.uint8)
    
    # Convert the contour array back to a SimpleITK image
    contour_image = sitk.GetImageFromArray(contour_array)
    contour_image.CopyInformation(liver_image)
    
    # Save the contour image
    contour_image_path = liver_image_path.replace('.nii.gz', '_contour.nii.gz')
    sitk.WriteImage(contour_image, contour_image_path)
    print(f"Liver contour image saved to {contour_image_path}")

# # Example usage
# liver_image_path = "/Users/wangzhengwei/Desktop/repo/3Dircadb1/3Dircadb1.2/liver_cropped_clip_resampled.nii.gz"
# extract_and_save_liver_contour(liver_image_path)
```

```python
def distance2weight(dist, gama=2):
    if type(dist) == torch.Tensor:
        dist= dist.numpy()

    weight = np.exp(-(dist**2)/(2*gama)) + 1

    return weight
```

```python
import os
import json
import random

def update_dataset_json(folder_root, json_path):
    # Check if the JSON file exists
    if not os.path.exists(json_path):
        raise FileNotFoundError(f"No such file or directory: '{json_path}'")
    
    #  JSON 
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    # 
    subdirs = [d for d in os.listdir(folder_root) if os.path.isdir(os.path.join(folder_root, d))]
    
    # 
    random.shuffle(subdirs)
    
    #  14:2:4 
    num_subdirs = len(subdirs)
    num_train = int(num_subdirs * 0.7)
    num_valid = int(num_subdirs * 0.1)
    
    training_subdirs = subdirs[:num_train]
    valid_subdirs = subdirs[num_train:num_train + num_valid]
    test_subdirs = subdirs[num_train + num_valid:]
    
    # 
    def construct_paths(subdirs):
        return [{'image': os.path.join(d, 'patient_cropped_clip_resampled.nii.gz'), 
                 'label': os.path.join(d, 'liver_cropped_clip_resampled.nii.gz'),
                 'distance': os.path.join(d, 'liver_cropped_clip_resampled_distance.nii.gz'),
                 'vessel_contour_distance': os.path.join(d, 'liver_contour_vessels_combined_distance.nii.gz'), 
                 'portalvein': os.path.join(d, 'portalvein_cropped_clip_resampled_modified.nii.gz'),
                 'venacava': os.path.join(d, 'venacava_cropped_clip_resampled_modified.nii.gz'),
                 'sato': os.path.join(d, 'patient_cropped_clip_resampled_sato.nii.gz')
                 } for d in subdirs]
    
    training_paths = construct_paths(training_subdirs)
    valid_paths = construct_paths(valid_subdirs)
    test_paths = construct_paths(test_subdirs)
    
    #  JSON 
    data['training']= training_paths
    data['valid'] = valid_paths
    data['test'] = test_paths
    
    #  JSON 
    with open(json_path, 'w') as f:
        json.dump(data, f, indent=4)

# 
json_path = './3dircab/3dircab_dataset.json'
update_dataset_json(folder_root, json_path)
```
